```r
## correction par application des fonctions de transfert au pas de temps mensuel

myqmaptn <- vector("list", length = 12)
ptm <- proc.time()
for(j in seq(along=myqmaptn)){
  myqmaptn[[j]] <- doQmapRQUANT(CMIP6Month[[j]], myqfit[[j]])
}
proc.time() - ptm
```

```r
###### Reconstruction de la sÃ©rie de Tn ou Tx ou RR

CMIP6TnSSPqmap <- do.call(rbind, myqmaptn)

library(xts)
CMIP6TnSSPcor.xts <- xts(CMIP6TnSSPqmap,order.by=as.Date(row.names(CMIP6TnSSPqmap)))

save(CMIP6TnSSPcor.xts,myqfit, file="../Data/CMIP6TnSSPcor.RData")
```

```r
load("../Data/lMFMonthTn.RData")
## compare les distributions
plot(density(CMIP6.xts[,1]),col="blue")
lines(density(MFTn.xts[,1]),col="red",lwd=4)
lines(density(CMIP6TnSSPcor.xts[,1]),col="green")
```
