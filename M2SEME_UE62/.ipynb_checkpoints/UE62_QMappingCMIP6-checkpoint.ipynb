```r
###########################################################################
library(qmap) ## package pour la correction quantile mapping (cf. Gudmundsson et al. 2012)

## création d'une fonction pour l'ajustement de la fonction de transfert Q-Q Mapping
myfqmapq <- function(x,y){
    ## Quantile mapping robuste (cf. Gudmundsson et al. 2012)
    qm.a.fit <- fitQmapRQUANT(x, y, qstep=0.01, nboot=10, wet.day=FALSE)
    return(qm.a.fit)
}
```

```r
##########################################################################
## ajustement de la fonction de transfert Quantile Mapping
## une fonction de transfert par mois et par Station-Point de grille modèle 
myqfit <- vector("list", length = 12)
ptm <- proc.time()
for(i in seq(along=myqfit)){
  myqfit[[i]] <- myfqmapq(x=lMFMonth.Tn[[i]], y=CMIP6Month[[i]])
}
proc.time() - ptm
```

```r
## correction par application des fonctions de transfert au pas de temps mensuel

myqmaptn <- vector("list", length = 12)
ptm <- proc.time()
for(j in seq(along=myqmaptn)){
    myqmaptn[[j]] <- doQmapRQUANT(CMIP6Month[[j]], myqfit[[j]])
}
proc.time() - ptm
```

```r
###### Reconstruction de la série de Tn ou Tx ou RR

CMIP6Tnqmap <- do.call(rbind, myqmaptn)

library(xts)
CMIP6Tncor.xts <- xts(CMIP6Tnqmap, order.by=as.Date(row.names(CMIP6Tnqmap)))

#save(CMIP6Tncor.xts, myqfit, file="../Data/FunqmapTn.RData")
```
